```python
import pandas as pd
df = pd.read_pickle('merged_data.pkl')
from datetime import datetime
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
from sklearn.feature_extraction import DictVectorizer
from sklearn import cross_validation, naive_bayes
import numpy as np
```

```python
df.head()
```

##ngram repetion t-test

```python
winter = [12,1,2]
summer = [6,7,8]
hot = []
cold = []
for person in df.ngram_frequency_month.iteritems():
    for key, value in person[1].iteritems():
        month = datetime.fromtimestamp(float(key)).date().month
        if month in winter:
            cold.append(value[0])
        elif month in summer:
            hot.append(value[0])
```

```python
paired_sample = ttest_ind(hot, cold)

print "The t-statistic is %.3f and the p-value is %.3f." % paired_sample
```

##gender predict

```python
import nltk
from nltk.corpus import stopwords
stops = stopwords.words('english')
def optimize_bag(bag):
    for key,value in bag.iteritems():
        if key in stops:
            del key
    return bag
```

```python
men = []
women = []
with open('genders.txt', 'r') as f:
    for line in f:
        if(line.split()[1] == 'm'):
            male.append(line.split()[0])
        elif(line.split()[1] == 'f'):
            female.append(line.split()[0])

for indx, row in df.iterrows():
    bag = optimize_bag(row['bag_of_words_total'])
    if row['gender'] is 'm':
        men.append({'male': True, 'bag': bag})
    elif row['gender'] is 'f':
        women.append({'male': False, 'bag': bag})
    else:
        print 'no gender!'
genders = pd.DataFrame.from_records(men + women)
```

```python
genders.head(12)
```

```python
v = DictVectorizer()
X = v.fit_transform(genders.bag)
X = X.tocsc()
Y = genders.male.values.astype(np.int)
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X,Y)
clf = naive_bayes.MultinomialNB(fit_prior=False, alpha=0.5)
clf.fit(X_train, Y_train)
print "Accuracy: %0.2f%%" % (100 * clf.score(X_test, Y_test))
```

```python
words = np.array(v.get_feature_names())

x = np.eye(X_test.shape[1])
probs = clf.predict_log_proba(x)[:, 0]
ind = np.argsort(probs)

male_words = words[ind[:10]]
female_words = words[ind[-10:]]

male_prob = probs[ind[:10]]
female_prob = probs[ind[-10:]]

print "Male words\t P(male | word)"
for w, p in zip(male_words, male_prob):
    print "%20s" % w, "%0.2f" % (1 - np.exp(p))
    
print "\nFemale words\t P(female | word)"
for w, p in zip(female_words, female_prob):
    print "%20s" % w, "%0.2f" % (1 - np.exp(p))
```

##bag of words naive bayes (comparing seasons)

```python
male = []
female = []
winter = [12,1,2]
summer = [6,7,8]
for month_bags in df.bag_of_words:
    for bag in month_bags.iteritems():
        month = datetime.fromtimestamp(float(bag[0])).date().month
        if month in winter:
            cold.append({'summer': False, 'bag': bag[1]})
        elif month in summer:
            hot.append({'summer': True, 'bag': bag[1]})
seasons = pd.DataFrame.from_records(hot + cold)
```

```python
seasons.tail()
```

```python
v = DictVectorizer()
X = v.fit_transform(seasons.bag)
X = X.tocsc()
Y = seasons.summer.values.astype(np.int)
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X,Y)
clf = naive_bayes.MultinomialNB(fit_prior=False, alpha=0.5)
clf.fit(X_train, Y_train)
print "Accuracy: %0.2f%%" % (100 * clf.score(X_test, Y_test))
```

```python
words = np.array(v.get_feature_names())

x = np.eye(X_test.shape[1])
probs = clf.predict_log_proba(x)[:, 0]
ind = np.argsort(probs)

summer_words = words[ind[:10]]
winter_words = words[ind[-10:]]

summer_prob = probs[ind[:10]]
winter_prob = probs[ind[-10:]]

print "Summer words\t P(summer | word)"
for w, p in zip(summer_words, summer_prob):
    print "%20s" % w, "%0.2f" % (1 - np.exp(p))
    
print "\nWinter words\t P(winter | word)"
for w, p in zip(winter_words, winter_prob):
    print "%20s" % w, "%0.2f" % (1 - np.exp(p))
```
